```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv("../data/train_delay_dataset.csv")

# Distribution
data["arrival_delay"].hist()
plt.title("Arrival Delay Distribution")
plt.show()

# Correlation
sns.heatmap(data.corr(), annot=True)
plt.show()

# Weather impact
data.groupby("weather")["arrival_delay"].mean().plot(kind="bar")
plt.title("Delay by Weather")
plt.show()

# Distance vs delay
plt.scatter(data["distance_km"], data["arrival_delay"])
plt.xlabel("Distance")
plt.ylabel("Delay")
plt.show()
```
